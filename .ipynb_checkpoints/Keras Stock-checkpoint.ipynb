```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfSPY = pd.read_csv('data/SPY.csv', index_col='Date',parse_dates=True, usecols=['Date','Adj Close'], na_values=['nan'])
```

```python
dfSPY = dfSPY.dropna()
```

```python
dfSPY.plot()
```

```python
dfSPY = dfSPY.sort_values(by=['Date'])

daily_returns = dfSPY.copy()
   
daily_returns[1:] = (dfSPY[1:] / dfSPY[:-1].values) - 1

daily_returns.iloc[0,:] = 0 # set daily returns for row 0 to 0
```

```python
daily_returns['Returns'] = daily_returns['Adj Close']
```

```python
daily_returns.plot()
```
